# Part 1

Assuming the file is in a CSV format, we can write a MapReduce program to solve this problem. We can process the file in two steps. 

## Step 1

In the first step, we can use MapReduce to find the largest value in each column. The input to this step will be the CSV file, where each row is a record. The output from this step will be a key-value pair, where the key is the name of the column, and the value is the largest value in that column.

The Mapper function will emit a key-value pair for each cell in the CSV file. The key will be the name of the column, and the value will be the cell value. The Reducer function will receive all the values for a given key (column name) and will emit a key-value pair where the key is the column name and the value is the largest value in that column.

## Step 2

In the second step, we can use MapReduce to find the smallest value in each row. The input to this step will be the CSV file, where each row is a record. The output from this step will be a key-value pair, where the key is the name of the row, and the value is the smallest value in that row.

The Mapper function will emit a key-value pair for each cell in the CSV file. The key will be the name of the row, and the value will be the cell value. The Reducer function will receive all the values for a given key (row name) and will emit a key-value pair where the key is the row name and the value is the smallest value in that row.

## Output

The final output will be of the form:  

"A", 564
"B", 329


There should be 20 lines in the report, and they do not have to be in any particular order.


```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file = open('cs4650hw1.dat', 'r')

def read_file(file):
    data = []
    for line in file:
        line = line.strip()
        line = line.split(',')
        data.append(line)
    return data

def create_df(data):
    # colNames = ['Column', 'Row', 'Value']
    df = pd.DataFrame(data, columns=['Column', 'Row', 'Value'])
    return df

def get_col_max(df):
    max_df = df.groupby('Column')['Value'].max()
    return max_df

def get_row_min(df):
    min_df = df.groupby('Row')['Value'].min()
    return min_df

def main():
    data = read_file(file)
    df = create_df(data)
    max_df = get_col_max(df)
    min_df = get_row_min(df)
    print(max_df)
    print(min_df)
if __name__ == '__main__':
    main()
```

```python
# now with a mapper and reducer
# %%file maxcol-minrow.py

# import re
from mrjob.job import MRJob

# QUOTE_RE = re.compile(r'["\']')

class MRMaxColMinRow(MRJob):

    def mapper(self, _, line):
        line = line.strip()
        line = line.split(',')
        yield line[0], int(line[2])

    def reducer(self, key, values):
        yield key, max(values) if key.isalpha() else min(values)
        
if __name__ == '__main__':
    MRMaxColMinRow.run()
```
